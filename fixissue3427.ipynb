```python
import qiskit
```

```python
import qiskit.pulse as pulse
import qiskit.pulse.pulse_lib as pulse_lib
sched = pulse.Schedule()
sched += pulse_lib.constant(10, 1.0)(pulse.DriveChannel(1))
sched.draw(channels_to_plot=[pulse.DriveChannel(1)])
```

```python
pulse_lib.gaussian(duration=10, amp=1, sigma=4).draw()
```

```python
sched += pulse_lib.gaussian(duration=10, amp=1, sigma=4)(pulse.DriveChannel(0))
sched.draw(channels_to_plot=[pulse.DriveChannel(0)])
```

```python
sched += pulse_lib.square(10, 1.0)(pulse.DriveChannel(2))
sched.draw(channels_to_plot=[pulse.DriveChannel(2)])
```
